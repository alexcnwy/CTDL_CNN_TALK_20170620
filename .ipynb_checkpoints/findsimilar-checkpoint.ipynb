```python
from __future__ import division

# setup matplotlib to display plots in the notebook
%matplotlib inline

# third party imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# setup display options
pd.options.display.max_rows = 200
pd.options.display.float_format = '{:,.2f}'.format
np.set_printoptions(precision=5, suppress=True)

# setup seaborn to use matplotlib defaults & styles
sns.set()
sns.set(font_scale=1.2)
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
plt.style.use('fivethirtyeight')

# setup some filepaths
pwd = !pwd
pwd = pwd[0]
```

# read CNN vectors for products

```python
prods = pd.read_csv(pwd + '/data/vectors/train.csv')
```

# read CNN vectors for new images

```python
test_vecs = pd.read_csv(pwd + '/data/vectors/test.csv')
test_vecs.head(100)
```

```python
test_idx = 4
test_vecs.loc[test_idx][0]
```

```python
test = test_vecs.loc[test_idx][1:]
test = test.reshape(len(test),1)
a = np.dot(prods.ix[:,range(1,prods.shape[1])], test)
results = pd.DataFrame(prods['Unnamed: 0'])
results['scores'] = a
results.sort_values('scores', ascending = False, inplace = True)
results.head()
```
